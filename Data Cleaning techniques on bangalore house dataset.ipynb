```python
#Filling null values using techniques
df2 = df.fillna(value = 0) # By creating duplicate data frame it will not effect on original data it was replaced by value 0.
df2
```

```python
df2.isnull().sum().sum()
```

```python
df3 = df.fillna(value = 5)
df3
```

```python
# filling null values by using the pervious values it is nothing but a forward fill
df4 = df.fillna(method = 'pad')  #df4 = df.fillna(method = 'ffill')
df4
```

```python
df4.isnull().sum().sum()
```

```python
#filling null values by using the next values it is nothing but a backward fill
df5 = df.fillna(method = 'bfill')
df5
```

```python
df5.isnull().sum().sum()
```

```python
#nearest fill (method='nearest'):
#Fills missing values with the nearest non-null value along the specified axis
df6 = df.fillna(method = 'pad', axis = 1) # It gives the columns wise to fill the null values
df6
```

```python
df6.isnull().sum().sum()
```

```python
df7 = df.fillna(method = 'bfill', axis = 1)
df7
```

```python
df7.isnull().sum().sum()
```

```python
df
```

```python
#Filling different values in null on different columns
df8 = df.fillna({'society' : 'Hope',    # by creating the dictionary we can fill the null values of the columns
                 'balcony' : 'sure'})
df8
```

```python
df8.isnull().sum().sum()
```

```python
#filling null values with the mean of the column
df9 = df.fillna(value = df['balcony'].mean())
df9
```

```python
df9 = df.fillna(value = df['balcony'].max())
df9
```

```python
df9 = df.fillna(value = df['balcony'].min())
df9
```

### Dropping method

```python
# Dropping method
df10 = df.dropna()
df10
```

```python
df10.isnull().sum().sum()
```

```python
# It means if the all the columns have null values it will dropped all otherwise one value colummn is null it is not dropped.
df11 = df.dropna(how = 'all') 
df11
```

```python
#If we have just null values in any places it will be dropped whole column.
df11 = df.dropna(how = 'any')
df11
```

```python
df11.isnull().sum().sum()
```

# Replace method

```python
#replacing the all null values with giving one input value
df12 = df.replace(to_replace=np.nan, value = 12345)
df12
```

```python
# we can also replace the existing values we can insert the new values at that place.
df13 = df.replace(to_replace=3.0, value = 6.0)
df13
```

```python
# we can also replace the existing values we can insert the new values at that place.
df14 = df.replace(to_replace='Chikka Tirupathi', value = 'Tirupathi')
df14
```

# Interpolation

```python
#interpolation (method='linear'):

#Performs linear interpolation to estimate missing values based on the values before and after the missing points.
```

```python
df15 = df['balcony'] = df['balcony'].interpolate(method = 'linear')
df15
```

```python
df15 = df['balcony'] = df['balcony'].interpolate(method = 'linear', limit_direction = 'forward')
df15
```

```python
# Mean Imputation
df16 = df.fillna(df.mean())
df16
```

```python
# Median Imputation
df_median = df.fillna(df.median())

# Mode Imputation (for categorical data)
df_mode = df.fillna(df.mode().iloc[0])

print("\nMedian Imputation:")
print(df_median)

print("\nMode Imputation:")
print(df_mode)
```
